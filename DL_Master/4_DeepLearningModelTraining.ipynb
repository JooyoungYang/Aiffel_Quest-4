## 4-2 Loss Function
손실함수란? 
정의)
- 학습이 진행되면서 해당 과정이 얼마나 잘 되고 있는지 나타내는 지표 <br>
- 모델이 훈련되는 동안 최소화될 값으로 주어진 문제에 대한 성공지표 <br>
- 손실 함수에 따른 결과를 통해 파라미터를 조정하여 학습이 진행 <br>
- 손실함수는 최적화 이론에서 최소화 하고자 하는 함수로 미분 가능한 함수를 사용 <br>
- Keras에서 제공되는 주요 손실 함수 <br>
    - 'Sparse_Categorical_Crossentropy': 클래스가 배타적 방식으로 구분, 즉 (0, 1, 2, ..., 9)와 같은 방식으로 구분되어 있을때 사용 <br>
    - 'Categorical_cross_entropy': 클래스가 원-핫 인코딩(one-hot encoding)방식으로 되어 있을 때 사용 <br>
    - 'binary_crossentropy': 이진 분류를 수행할 때 사용 <br>

1. 평균절대오차 <br>
    - 오차가 커져도 손실함수가 일정하게 증가 <br>
    - 이상치(Outlier)에 강건한(Robust)특성 <br>
    - 데이터에서 (입력, 정답) 관계가 적절하지 않은 것이 있을 경우에, 좋은 추정을 하더라도 오차가 발생하는 경우가 발생하는데, 이상치에 해당하는 지점에서 손실함수의 최소값으로 가는 정도의 영향력이 크지 않은 특성을 가지고 있음 <br>
    - 일반적으로 회귀(Regression)에 많이 사용하는 손실함수 <br>

2. 평균제곱오차 (Mean Squared Error, MSE) <br>
    - 가장 많이 사용되는 손실 함수 중 하나 <br>
    - 오차가 커질수록 손실함수가 빠르게 증가하는 특성 <br>
    - 정답과 예측한 값의 차이가 클수록 더 많은 페널티를 부여하는 형태로 동작 <br>
    - 일반적우로 회귀에 사용되는 손실함수 <br>

3. 원-핫 인코딩 <br>

4. 교차 엔트로피 오차

5. 이진 분류 문제의 교차 크로스 엔트로피

## 4-3 옵티마이저와 지표
정의)
- 손실함수를 기반으로 모델이 어떻게 업데이트되어야 하는지 결정
- Keras에서 여러 옵티마이저를 제공하고, 사용자가 특정 종류의 확률정 경사하강법 지정 가능
    - 'keras.optimzer.SGD()': 기본적인 확률적 경사 하강법 <br>
    - 'keras.optimizer.Adam()': 자주 사용되는 옵티마이저 <br>
- 보통 옵티마이저의 튜닝을 위해 따로 객체를 생성하여 컴파일시에 포함

1. 경사하강법(Gradient Decent)
    - 미분과 기울기로 동작하며, 스칼라를 벡터로 미분 <br>
    - 변화가 있는 지점에서는 미분값이 존재하고, 변화가 없는 지점은 미분값이 0이 되며, 미분값이 클수록 변화량이 큼 <br>
    - 경사하강법의 과정은 한 스텝마다의 미분값에 따라 이동하는 방향을 결정, f(x)의 값이 변하지 않을때까지 반복 <br>
    - learning_rate <br>
노드 Q) 경사하강법은 무슨 일을 하나요? <br>
A) 옵티마이저는 손실 함수를 기반으로 모델이 어떻게 업데이트되어야 하는지 결정합니다. Keras에서 여러 옵티마이저를 제공하고, 사용자는 특정 종류의 확률적 경사 하강법을 지정할 수 있습니다. <br>
노드Q) 학습률은 어디에 사용되고 잘못 설정한 경우 어떤 문제가 발생할 수 있나요? <br>
A) 경사하강법을 통해 손실함수의 반대 기울기 방향으로 업데이트 할 때 업데이트 되는 크기에 관여하며 잘못 설정한 경우 손실함수의 최저점에 도달하지 못할 수 있습니다. <br>

2. 볼록함수와 비볼록함수 (Convex Function & Non-Convection Function)
     - 복록함수는 어떤 지점에서 싲가하더라도 최적값(손실함수가 최소로하는 점)에 도달가능 <br>
     - 비볼록함수는 시작점 위치에 따라 다른 최적값에 도달할 수 있음 <br>
3. 안장점
    - 기울기가 0이지만 극값이 되지 않는 안장점이 존재 <br>
    - 경사하강법은 안장점에서 벗어나지 못하는 문제가 있음 <br>
노드Q) 왜 경사하강법으로는 안장점(Saddle point)에서 벗어나지 못할까요?
A) 기울기의 반대 방향으로 다음 위치를 이동해야하는데 기울기 값이 0이라 이동하지 못합니다.
    
4. 학습률 (LearningRate)
    - 모델을 학습하기 위해서는 적절한 학습률을 지정해야 최저점에 잘 도달할 수 있음 <br>
    - 학습률이 너무 크면 발산하고, 너무 작으면 학습이 오래 걸리거나 최저점에 도달하지 못하는 문제가 있음 <br>
5. 지표 (Metrics)
    - 딥러닝 학습 시 필요한 다양한 지표들을 지정가능
    - 일반적으로 'mae'나 'accuracy'를 사용
    - 'accuracy'와 같은 경우 줄여서 'acc'로도 사용이 가능

#### 데이터 생성

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split


X, y = make_regression(n_samples=200, n_features=1, 
                       bias=5.0, noise=5.0, random_state=123)
y = np.expand_dims(y, axis=1)

plt.scatter(X, y)
plt.show()


x_train, x_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, 
                                                    shuffle=True,
                                                    random_state=123)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

#### 모델 생성

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import models, layers, optimizers, utils
```

```python
model = keras.Sequential()
model.add(layers.Dense(1, activation='linear', input_shape=(1,)))
model.summary()
```

```python
utils.plot_model(model)
```

#### 모델 학습

```python
# SGD(Stochastic Gradient Descent) 

optimizer = optimizers.SGD()
model.compile(loss='mse', optimizer=optimizer, metrics=['mae', 'mse'])
history = model.fit(x_train, y_train, epochs=40)
```

```python
plt.plot(history.history['mae'])
plt.plot(history.history['mse'])
plt.xlabel('Epoch')
plt.legend(['mae', 'mse'])
plt.show();
```

#### 모델 평가

```python
model.evaluate(x_test, y_test)
```

#### 모델 예측

```python
result = model.predict(X)

plt.scatter(X, y)
plt.plot(X, result, 'r')
plt.show()
```

```python
plt.scatter(X, y)
plt.plot(X, np.array(weights * X + biases), 'r')
plt.show()
```
