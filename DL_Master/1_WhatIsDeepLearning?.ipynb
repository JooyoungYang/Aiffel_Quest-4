## 1-3 ArtificialNeuralNetwork
01인공신경망정의
- 인간의 뇌 구조를 모방: 뉴련과 뉴런 사이의 전기신호를 통해 정보를 전달
- 인간의 뉴런 구조와 활성화 작동원리를 input(자극)과 output(반응)과의 연관으로 구현 <br><br>

02신경세포

- 축색돌기

03 인공뉴런

- 활성화함수(activation function)를 통한 비선형구조(non-linear)로 표현 가능 <br><br>

04 인공신경망구조

- 은닉층 (hidden layer)
- 출력층 (output layer)

## 1-4 인공신경망 역사
01최초의인경망
- 인간의 신경계를 활성상태화 비활성 상태를 갖는 이진 뉴런으로 표현
- 생체 뉴런과 시냅의 흥분과 억제에 따라 신호가 전달되고 특정 임계치가 넘어야 신호가 발화 <br><br>

02 퍼셉트론Perceptron
- 프랭크 로젠블랫은 인공신경망이 스스로 문제에 맞춰 학습하는 퍼셉트론 개발
- 생체신경망을 모방해서 만든 퍼셉트로능의 처리과정과 뉴런의 신호 전달과정을 비교

- 마빈민스키와 시모어 페퍼트는 퍼셉트론 비선형 문제를 해결할 수 없음을 증명( XOR 논리연산증명)
- 퍼셉트론 2개 이상의 직선을 표현하려명 AND 연산 필요, 여러 계층으로 표현되는 다층 퍼셉트론(MLP) 탄생 

노드 Q. 영상에 등장하진 않았지만 다른 논리회로 게이트 종류도 많습니다. 다른 논리 회로 게이트 중에서도 단일 퍼셉트론으로는 해결할 수 없는 것들이 있을까요? <br>
A. XOR 게이트의 반전 형태인 XNOR 게이트가 있습니다. XNOR 게이트는 입력 받은 데이터가 같으면 1, 다르면 0을 출력합니다. <br><br>


03 다층퍼셉트론Multi-LayerPerceptron(MLP) 
- 여러개의 퍼셉트론을 층으로 구성한 신경망 모델
- 다층 퍼셉트론을 학습하기 어려운 것은 활성화함수가 계단함수라 미분이 불가능하여 미분가능한 함수 필요 
노드Q. 다층 퍼셉트론을 학습시키는것이 어려운 이유는 무엇일까요?
A. 많은 은닉층의 수와 미분이 불가능한 계단함수의 존재가 학습의 난이도를 높임. <br><br>

04 역전파 알고리즘(Backpropagation)
- 역전파 알고리즘은 1982년에 폴 워보스가 학위논문으로 처음 발표했고, 이후 1985년에 데이비드 파커와 Yann LeCun이 제안, 1986에 David Rumelhart, James McCleeland, Geoffrey Hinton, Ronald Williams가 공개 

## 1-5 딥러닝 역사
01인공신경망과 딥러닝
- 일반적으로 인공신경망은 은닉층이 소수인 반면, 딥러닝은 은닉층이 다수포함된 심층신경망(deep neural network)형태
- 일반신경망: 데이터에서 직접특징(handcrafted feature)을 추출해서 만든 특징벡터(feature vector)를 입력으로 사용하고, 특징 벡터들의 품질에 영향
 <br>
특징추출을 포함했을때 장점)사람의 개입이 들어가지 않아 사람의 편견으로 인한 오류가 적고 추론시간이 짧아지며 성능이 높아짐 

- 딥러닝신경망: 특징추출과 학습을 함께 수행하고, 데이터로부터 효과적인 특징을 학습을 통해 추출하여 우수한 성능 달성
 <br>

- 장점) 높은 성능과 뛰어한 확장성으로 이미지 처리, 자연어 처리, 음성 인식과 같은 분야에서 급격한 발전이 이루어지고 있음
- 단점) 딥러닝 모델은 필요한 파라미터(parameter)수가 많아서 더 많은 학습데이터가 필요하고, 최적의 모델과 훈련을 위해 많은 시간과 비용이 필요 <br>

#### 딥러닝에서 발생할 수 있는 문제 <br>
기울기소멸문제: 
- 정의) 
- 해결방법)시그모이드나 쌍곡탄젠트대신 ReLU(Rectified Linear Unit)이라고 부름 <br>

과적합: 
- 정의) 과도하게 학습, 학습하지 않은 데이터에 대해서는 오히려 성능이 저하 <-> 과소적합(Underfitting)은 모델이 작거나 단순해서 데이터를 잘 표현하지 못한 상태로 모델의 크기를 키우거나 좋은 모델로 변경필요

- 해결방법) 
1. 규제화(Regularization)기법: 큰 가중치 값에 큰 규제를 가하여 과적합되지 않도록 모델을 제한, 규제의 강도를 정하는 적절한 가중치가 필요
2. 드롭아웃(Dropout)기법: 노드를 일정확률로 무작위 선택하여 연결된 가중치 연결선을 없는 것으로 간주하고 학습하는 방법
3. 배치정규(Batch Normalization)기법: 모델이 입력되는 샘플들을 균일하게 만드는 방법으로 미니 배치 단위로 평균이 0, 표준쳔차가 1이 되도록 정규화

딥러닝의 발전
- 제프리 힌튼
- 조슈아 벤지오
- 얀 르쿤

Framework
- TensorFlow:
    - 구글이 개발
    - 파이썬, C++ API, JavaScript, Java, Go, Swift등 다양한 언어 지원 (파이썬만 지원하는 줄 알았는데 요건 몰랐음)
- Keras
- PyTorch
    - Facebook에서 Lua라는 언어로 개발된 Torch를 파이썬을 지원하는 형식으로 만듦

딥러닝은 계속발전:
- 전이학습(Transfer Learning)
- 자기지도학습(Self-supervised Learning)
- 메타학습(Meta Learning): 학습방식을 학습(learn to learn)하여 인간과 같이 적은 데이터로도 빠르게 학습하는 방법
- 설명가능한 인공지능(Explainable AI, XAI): 신경망 모델의 내부나 동작 원리를 파악할 수 있도록 하는 기법
- Neural Architecture Search, NAS: 최적의 딥러닝 모델을 자동으로 찾아주는 방법
- Auto ML: 강화학습 방식으로 최적의 딥러닝 모델 생성

- 생체신경망을 모방해서 만든 퍼셉트론과 다층퍼셉트론 그리고 역전파 알고리즘
- 인공신경망은 은닉층이 소수인 반면, 딥러닝은 은닉층이 다수 포함된 심층신경망
- 딥러닝의 문제점인 기울기 소멸문제와 과적합 문데를 해결하여 계소해서 발전
